### 1.3 Assignment Operators

In python, the operator `=` has a special meaning.
Consider the following code:
` c = a + b `

Python will first compute the expression on the right hand side of the operator `=`,
followed by assigning the value to the variable `c`.

For example,
The code ` a = a + b` may make no sense to you. However, for python, it makes much sense.

Python will compute the value of `a + b`, then assign it back to the variable `a`.

```python
# example:
a = 3
b = 4
a = a + b
print(a)
```

Python also supports some short form of assignment operators, such as:

| Operator | Example | Equivalent to |
| - | - | -|
| += | a += b | a = a + b |
| -= | a -= b | a = a - b |
| *= | a *= b | a = a * b |
| /= | a /= b | a = a / b |
| %= | a %= b | a = a % b |
| **= | a **= b | a = a ** b |


```python
# example:
a = 3
b = 4
a += b
print(a)
```

### 1.4 Python Comment

In Python, comment is used to explain your code and your logic.

Comment starts with `#`, and Python will not execute the code mark beyond the `#` operator in the same line.

Even though the comment code will not be executed, you should always provide comments in your code to explain your logic to others. To make your program more readable.



```python
# this is a comment, and python will not execute it
a = a + b #this is also a comment, python will execute the code a = a + b but not the code beyond the comment operator.
print(a)
```

### 1.6 Data Type Conversion

You can cast a variable into another data type using the casting function. Here're the 3 most common casting functions used in Python.

<code>
    int()
    float()
    str()
</code>
    

```python
x = 1.8      # x is a float number
y = "1.8"    # y is a string

a = int(x) # x will become integer, it will round down to 1
b = str(x) # x will become a string, the string is stored in variable b3
c = float(y) # y will become a float number.

print([a,b,c]) # I form a list for return the two new data in one line
```
