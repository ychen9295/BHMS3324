A function in a programming language is a block/sequence of program codes to perform a defined task.

### Example 1:

Define a greeting function `say_hello()`:

In Python, a function is defined using the `def` keyword. The parenthesis `()` tells Python that you're declaring a function, and you need to type `:` at the end of your function.

```python
def say_hello():
    print("Hello, Raymond!")
```

You can call the function by its function name followed by parenthesis.

```python
say_hello()
```

You can also call it multiple times.

```python
say_hello()
say_hello()
say_hello()
```

```python
# let's say hi to me in 10 times

for i in range(10):
    say_hello()
```

You can also create a personalized greeting function by specifying parameters (arguments) in the function. You can set the arguments after the function name and inside the parentheses.

```python
# Hello with argument
def say_hello(name):
    print("Hello " + name)
```

```python
say_hello("Peter")
```

```python
# create a list of names
# greet them one by one

names = ["Joe", "Peter", "Bill", "Kris", "Isaac"]
for name in names:
    say_hello(name)
```

We can also specify default values for arguments if we call the function with none.

```python
# Hello with a default argument
def say_hello(name = "mate"):
    print("Hello " + name + "!")
```

```python
say_hello()
say_hello("Kate")
```

# Python Modules and Package

A module in Python is a code library. You can treat it as a file that contains a set of functions.
In Python, a package is a collection of Python modules, while a module is a single Python file that contains a set of functions.

In the lecture, we talk about the SHA256 hash algorithm, the core of Blockchain. In Python, the SHA256 function can be called after including the module in your program.


```python
import hashlib

input_message = "Hello"
output_hash = hashlib.sha256(input_message.encode()).hexdigest()

print(output_hash)
```
